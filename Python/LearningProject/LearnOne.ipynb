```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt


data = {
    'Book': ['The Hobbit', '1984', 'Pride and Prejudice', 'Dune', 'The Great Gatsby', 'To Kill a Mockingbird'],
    'Genre': ['Fantasy', 'Dystopian', 'Romance', 'Sci-Fi', 'Fiction', 'Fiction'],
    'Price': [15.99, 12.50, 10.00, 18.75, 9.99, np.nan],
    'Quantity_Sold': [50, 75, 30, 45, 60, 55],
    'Date': ['2023-01-01', '2023-01-01', '2023-01-02', '2023-01-02', '2023-01-03', '2023-01-03']
}

df = pd.DataFrame(data)

# df["Date"]=pd.to_datetime(df["Date"],format="mixed")
df['Date'] = pd.to_datetime(df['Date'])
df
df["Price"]=df["Price"].fillna(df["Price"].mean())
df["Revenue"]=df["Price"]=df["Price"] * df["Quantity_Sold"]
genre_summary=df.groupby("Genre").agg({"Quantity_Sold":'sum',"Revenue":'sum'}).reset_index()
plt.figure(figsize=(8,5))
plt.bar(genre_summary['Genre'],genre_summary['Revenue'],color='orange')
plt.title('Total Revenue By Genre')
plt.xlabel('Genre')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.plot()


sales_by_date=df.groupby('Date')['Quantity_Sold'].sum().reset_index()
plt.figure(figsize=(8,5))
plt.plot(sales_by_date['Date'], sales_by_date['Quantity_Sold'], marker='o', color='purple')
plt.title('Total Sales By Date')
plt.xlabel('Date')
plt.ylabel('Quantity Sold')
plt.grid(True)
plt.show()


## scater plot 
sales_by_date = df.groupby('Date')['Quantity_Sold'].sum().reset_index()
plt.figure(figsize=(8,5))
plt.title('Plotting using Scatter Plot')
plt.xlabel('Date')
plt.ylabel('Quantity Sold')
plt.scatter(sales_by_date['Date'], sales_by_date['Quantity_Sold'], marker='o', color='green')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

## pie chart 
sales_by_date = df.groupby('Date')['Quantity_Sold'].sum().reset_index()

plt.figure(figsize=(6,6))
plt.pie(sales_by_date['Quantity_Sold'],labels=sales_by_date['Date'].dt.strftime('%Y-%m-%d'),  autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral', 'lightgreen'])
plt.title('Proportion of Quantity Sold by Date')
plt.axis('equal') 
plt.show()



```
