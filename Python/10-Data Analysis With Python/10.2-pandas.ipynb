# Pandas is a powerful Python library for data manipulation and analysis.
# It provides data structures like Series (1D) and DataFrame (2D) for handling and analyzing structured data efficiently.

```python
import pandas as pd
data=[1,2,3,4]
series=pd.Series(data)
series
```

```python
## Create a series from dictionary
data={'a':1,'b':2,'c':3}
series_dict=pd.Series(data)
print(series_dict)
```

```python
data=[10,20,30]
index=['a','b','c']
pd.Series(data,index=index)
```

```python
## Dataframe
## Create a Dataframe from dictionary of list

data={
    'Name':['Himanshu','john','jack'],
    'Age':[23,34,55],
    'city':['Banglore','newYork','florida']
}
df=pd.DataFrame(data)
print(df)
print(type(df))
```

```python
## create a dataframe from list
data=[
    {'Name':'Himanshu','age':24,'city':'banglore'},
     {'Name':'prince','age':22,'city':'varansi'},
      {'Name':'divyanshu','age':24,'city':'nepal'}
]
df=pd.DataFrame(data)
print(df)
```
